```python
prompt=prompts['prompt'].tolist();
prompt_type=prompts['source'].tolist();
```

```python
#converts lists of int into lists of str
ids_str=[]
for i in range(len(prompt)):
  ids_str.append(str(ids_range[i]))
```

```python
#create dictionary key=index and value=name of prompt for metadata
prompt_type_string_dict=[]
i=0
for i in range(len(prompt)):
  prompt_type_string_dict.append({ids_str[i]:prompt_type[i]})
```

```python
import chromadb
from chromadb.utils import embedding_functions
chroma_client = chromadb.Client()
```

**--------------------------------**

```python
#to delete a collection
collection = chroma_client.delete_collection(name="collection_of_known_prompts")
```

**--------------------------------**

```python
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-mpnet-base-v2")
collection = chroma_client.create_collection(name="collection_of_known_prompts", embedding_function=sentence_transformer_ef, metadata={"hnsw:space": "cosine"})
# vector database
```

```python
collection.add(
    documents=prompt,
    metadatas=prompt_type_string_dict,
    ids=ids_str
)
```

```python
#collection.peek() returns a list of the first 10 items in the collection
collection.count() 
```

```python
results=collection.query(
    query_texts=["how to eat?"],
    n_results=5
    #include=['distances','metadatas','documents']
)
```

**Similarity between prompts**

```python
simi1=[]
simi2=[] #list of each prompt when compared with every other prompt
for i in range(len(prompt)):
    simi1=[]
    for j in range(len(prompt)):
        simi1.append(1-distance.cosine(prompt_embeddings[i], prompt_embeddings[j]))
    simi2.append(np.mean(simi1))
```

```python
print(np.mean(simi2))
```
