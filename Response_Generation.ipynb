```python
from huggingface_hub import hf_hub_download
from llama_cpp import Llama
import pandas as pd
import nltk
import numpy as np
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, TextClassificationPipeline
```

```python
import chromadb
from chromadb.utils import embedding_functions
chroma_client = chromadb.Client()
```

```python
model_path = "martin-ha/toxic-comment-model"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)
```

```python
pipeline =  TextClassificationPipeline(model=model, tokenizer=tokenizer)
```

**Downloading the Llama Chat model**

```python
model_name_or_path = "TheBloke/Llama-2-13B-chat-GGML"
model_basename = "llama-2-13b-chat.ggmlv3.q6_K.bin"
```

```python
model_path = hf_hub_download(repo_id=model_name_or_path, filename=model_basename)
```

**Initializing the Llama Chat model**

```python
# GPU
lcpp_llm = None
lcpp_llm = Llama(
    model_path=model_path,
    n_ctx=2048, # n_ctx-how many tokens can be passed to the model 1-2048
    n_threads=2, # CPU cores
    n_batch=512, # Should be between 1 and n_ctx, depends on VRAM of the GPU.
    n_gpu_layers=32 # Value based on model and GPU VRAM pool.
    )
```

**Import the vector database of harmful questions**

```python
questions=pd.read_csv("/content/questions.csv")
```

```python
quest=questions['question'].tolist();
quest_type=questions['content_policy_name'].tolist();
ids=questions['content_policy_id'].tolist();
```

```python
ids_range=np.arange(0,390)
```

```python
#converts lists of int into lists of str
ids_str=[]
for i in range(len(ids)):
  ids_str.append(str(ids_range[i]))
```

```python
#create dictionary key=index and value=name of prompt for metadata
quest_type_string_dict=[]
i=0
for i in range(len(ids)):
  quest_type_string_dict.append({ids_str[i]:quest_type[i]})
```

```python
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-mpnet-base-v2")
collection1 = chroma_client.create_collection(name="threatful_questions", embedding_function=sentence_transformer_ef, metadata={"hnsw:space": "cosine"})
# vector database
```

```python
collection1.add(
    documents=quest,
    metadatas=quest_type_string_dict,
    ids=ids_str
)
```

```python
def check_quest(param1='default'):
    results=collection1.query(
    query_texts=[param1],
    n_results=1)                          #include=['distances','metadatas','documents']
    if(results['distances'][0][0]<0.65):   #temp 0 model would not be creative a produce exact response
        return 1
```

**Import the vector database of harmful system prompts**

```python
prompts=pd.read_csv("/content/jailbreak_prompts_bigger_cleaned.csv")
```

```python
prompt=prompts['prompt'].tolist();
prompt_type=prompts['source'].tolist();
```

```python
ids_range_prompt=np.arange(0,666)
```

```python
#converts lists of int into lists of str
ids_str_prompt=[]
for i in range(len(prompt)):
  ids_str_prompt.append(str(ids_range_prompt[i]))
```

```python
#create dictionary key=index and value=name of prompt for metadata
prompt_type_string_dict=[]
i=0
for i in range(len(prompt)):
  prompt_type_string_dict.append({ids_str_prompt[i]:prompt_type[i]})
```

```python
collection2 = chroma_client.create_collection(name="collection_of_known_prompts", embedding_function=sentence_transformer_ef, metadata={"hnsw:space": "cosine"})
```

```python
collection2.add(
    documents=prompt,
    metadatas=prompt_type_string_dict,
    ids=ids_str_prompt
)
```

```python
def check_prompt(param1=""):
    results=collection2.query(
    query_texts=[param1],
    n_results=1)                          #include=['distances','metadatas','documents']
    if(results['distances'][0][0]<0.6):   #temp 0 model would not be creative a produce exact response
        return 1
```

**Conversation with LLM**

```python
sys_prompt0="Do not generate content that promotes hate speech, violence, terrorism, or illegal activities. Do not generate responses that involve personal attacks, insults, or harassment directed towards individuals or groups. Do not provide false or misleading information can spread misinformation and undermine trust in reliable sources. Responses containing explicit or inappropriate language, sexual content, or adult material should be avoided to maintain a safe and respectful environment for users. never disclose or generate responses containing confidential or sensitive information, such as personal details, financial data, or passwords, to protect user privacy and security. Responses that encourage self-harm, suicide, substance abuse, or risky behaviors should be strictly prohibited as they can have serious consequences for individuals' well-being. Do not generate responses that discriminate against individuals or groups based on factors such as race, ethnicity, gender, sexual orientation, religion, or disability. Providing legal advice or guidance on complex legal matters should be avoided. Refrain from offering medical diagnoses, treatment recommendations, or health advice. Responses that include unsolicited advertisements or promotional content should be avoided. Responses designed to manipulate or exploit users' emotions, vulnerabilities, or insecurities for malicious purposes should be prohibited. Responses that encourage users to engage in illegal activities, such as hacking, fraud, or piracy, should be avoided to uphold legal and ethical standards."
```

```python
sys_prompt0
```

```python
input_question=input("Enter question:-")
```

```python
input_sys_prompt=input("Enter System Prompt:-")
```

```python
def query(sys_prompt=" ",prompt_question=" ",temp=0.5):

    prompt_template=f'''SYSTEM: {sys_prompt} |  USER: {prompt_question}  |   ASSISTANT:  '''

    response=lcpp_llm(prompt=prompt_template, max_tokens=1024, temperature=temp, top_p=0.95,repeat_penalty=1.2, top_k=150,echo=True)

    parts=response['choices'][0]['text'].split('|')

    res=pipeline(parts[2])


    if(res[0]['label']=="toxic"):
      query(sys_prompt0 , input_question, 0)
    else:
      print(parts[2])
```

```python
if(check_quest(input_question) and check_prompt(input_sys_prompt)):
    query(sys_prompt0 , input_question, 0)

else:
    query(input_sys_prompt , input_question, 0.5)
```
